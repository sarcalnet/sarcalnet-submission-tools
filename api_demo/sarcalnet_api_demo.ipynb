```python
geoDB.get_collection_info("calibration_sites", database="sarcalnet")
```

```python
geoDB.get_collection("calibration_sites", database="sarcalnet")
```

```python
geoDB.get_collection_pg("calibration_sites",
                        select="short_site_id,geometry,primary_target_type_id,"
                               "primary_sensor,landcover",
                        where="primary_target_type_id LIKE '%AT%'",
                        database="sarcalnet")
```

```python
geoDB.get_collection_pg("calibration_targets",
                        select="site_name,geometry,primary_direction",
                        where="(site_name = 'El Sosneado' or site_name = 'Piramides') and "
                              "upper(primary_direction) = 'EAST'",
                        database="sarcalnet")
```

```python
geoDB.get_collection_pg("calibration_surveys",
                        select="array_agg(target_id),survey_date",
                        where="survey_date >= '2024-01-01'::date and survey_date <= '2024-12-31'::date",
                        group="survey_date",
                        database="sarcalnet")
```
